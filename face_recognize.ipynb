```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, BatchNormalization
from tensorflow.keras.optimizers import Adam
from keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

TRAINING MODEL

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    brightness_range=[0.8, 1.2]
)

train_generator = train_datagen.flow_from_directory(
    "dataset",
    target_size=(128, 128),
    batch_size=2,
    class_mode='categorical'
)

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.1),

    Conv2D(64, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.1),

    Conv2D(128, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.1),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.1),
    Dense(3, activation='softmax')
])

model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

model.summary()

history = model.fit(
    train_generator,
    # batch_size=train_generator.samples // train_generator.batch_size,
    epochs=100,  
    callbacks=[early_stopping], 
)

model.save('face_recognition_model.h5')
```

DETECTION AND RECOGNIZE

```python
import cv2
import numpy as np
from PIL import Image
from mtcnn import MTCNN
from tensorflow.keras.models import load_model

model = load_model('face_recognition_model.h5')

detector = MTCNN()

def preprocess_image(img):
    img = Image.fromarray(img).resize((128, 128))  
    img_array = np.array(img) / 255.0 
    img_array = np.expand_dims(img_array, axis=0)  
    return img_array

def detect_and_classify_face(frame):
    detections = detector.detect_faces(frame)
    results = []

    for detection in detections:
        x, y, width, height = detection['box']
        x, y = abs(x), abs(y)
        x2, y2 = x + width, y + height

        face_img = frame[y:y2, x:x2]

        face_img_preprocessed = preprocess_image(face_img)

        prediction = model.predict(face_img_preprocessed)
        class_idx = np.argmax(prediction, axis=1)[0]

        class_names = ['Long','Lu','Minh'] #label
        class_name = class_names[class_idx]
        
        results.append((class_name, (x, y, x2, y2)))

    return results

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = detect_and_classify_face(frame)

    for class_name, (x1, y1, x2, y2) in results:
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, class_name, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    
    cv2.imshow('Face Detection and Classification', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
